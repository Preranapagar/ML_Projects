### 7.Plot each categorical feature

```python
for col in df.select_dtypes(include='object'):
    if df[col].nunique()<10:
        sns.countplot(y=col,data=df)
        plt.show()
```

### 8.Segment the variable by categorical features

```python
for col in df.select_dtypes(include='object'):
    if df[col].nunique()<10:
        sns.boxplot(y=col,x='Price',data=df)
        plt.show()
```

### 9. Group numeric features by each categorical feature

```python
for col in df.select_dtypes(include='object'):
    if df[col].nunique()<10:
        display(df.groupby(col).mean())
```

### 10. Correlations matrix for different numerical features

```python
corrs=df.corr()
corrs
```

### 11. Heatmap to visualize correlations

```python
plt.figure(figsize=(10,10))
sns.heatmap(corrs,annot=True)
plt.show()
```
